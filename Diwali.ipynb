```python
ax=sns.countplot(x='Gender',data=df,hue='Gender',palette='plasma')
def data_label(var):
  for i in var.containers:
    var.bar_label(i,label_type='center')
data_label(ax)
plt.show()
```

```python
a=df.groupby('Gender')['Amount']
female_spend=a.get_group('F').sum()
male_spend=a.get_group('M').sum()
print(f"The total spent of a female {female_spend}")
print(f"The total spent of a male {male_spend}")
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum()
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum()
d=sns.barplot(x='Gender',y='Amount',data=sales_gen,hue='Gender',palette='plasma')
data_label(d)
plt.show()
```

From the above graph we can see that most of the buyers are females and even purchasing power of females is greater than males and more number of revenue has been generated by females only.

State

```python
state=sns.countplot(x='State',data=df,hue='State',palette='plasma')
plt.xticks(rotation=90)
for i in state.containers:
  state.bar_label(i)
plt.show()
```

```python
orders_state=df.groupby('State',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
```

```python
state=sns.barplot(y='State',x='Orders',data=orders_state,hue='State',palette='plasma',legend=True)
plt.xticks(rotation=90)
for i in state.containers:
  state.bar_label(i)
plt.show()
```

```python
sales_state=df.groupby('State',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
state=sns.barplot(y='State',x='Amount',data=sales_state,hue='State',palette='plasma')
plt.xticks(rotation=90)
for i in state.containers:
  state.bar_label(i)
plt.show()
```

```python
sales_state=df.groupby('State',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
state=sns.barplot(y='State',x='Amount',data=sales_state,hue='State',palette='plasma')
plt.xticks(rotation=90)
for i in state.containers:
  state.bar_label(i)
plt.show()
```

From the above graph we can see that unexpectedly most of the orders from the uttar pradesh , maharastra and karnataka reespectively.
